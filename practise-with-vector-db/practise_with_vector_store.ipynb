```python
from langchain_huggingface import HuggingFaceEmbeddings
from google.colab import userdata
import os
os.environ['HF_TOKEN']=userdata.get('HF_TOKEN')

hf_embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
# from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

load_text = TextLoader("/content/speech.txt").load()
split_docs = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200).split_documents(load_text)

```

```python
db = FAISS.from_documents(split_docs, hf_embeddings)
```

```python
query = "How does the speaker describe the desired outcome of the war?"
db.similarity_search(query)[:1][0].page_content
```

```python
retriver = db.as_retriever()
retriver.invoke(query)[0].page_content
```

```python
emb_query = hf_embeddings.embed_query(query)
db.similarity_search_by_vector(emb_query)[0].page_content
```

```python
### Saving And Loading
db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",hf_embeddings,allow_dangerous_deserialization=True)
docs=new_db.similarity_search(query)
```

```python
docs[0].page_content
```

```python
chroma_db.similarity_search_by_vector(emb_query)[0].page_content
```
